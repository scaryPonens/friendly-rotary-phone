## NetworkX machine visualization


```python
G = machine_to_networkx(bubble_man_machine())
pos = nx.spring_layout(G, seed=7)
edge_labels = nx.get_edge_attributes(G, "label")

plt.figure(figsize=(8, 5))
nx.draw_networkx_nodes(G, pos, node_size=1700)
nx.draw_networkx_edges(G, pos, arrows=True, arrowstyle="->", arrowsize=20)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)
plt.axis("off")
plt.show()
```

## JSON schema export/import roundtrip


```python
struct_doc = struct_to_dict(struct)
reloaded_struct = struct_from_dict(struct_doc)
len(struct_doc["primitives"]), len(reloaded_struct.primitives), len(reloaded_struct.links)
```

```python
machine_doc = machine_to_dict(bubble_man_machine())
reloaded_machine = machine_from_dict(machine_doc)
reloaded_machine.start_state, sorted(reloaded_machine.accept_states)
```

```python
class_doc = class_representation_to_dict(bubble_man_class_representation())
reloaded_class = class_representation_from_dict(class_doc)
sorted(reloaded_class.constraints.keys())
```

## NetworkX struct visualization


```python
S = struct_to_networkx(struct)
pos = nx.spring_layout(S, seed=3)
labels = nx.get_node_attributes(S, "label")

plt.figure(figsize=(7, 4))
nx.draw_networkx_nodes(S, pos, node_size=1400)
nx.draw_networkx_edges(S, pos, arrows=True, arrowstyle="->", arrowsize=20)
nx.draw_networkx_labels(S, pos, labels=labels)
plt.axis("off")
plt.show()
```
